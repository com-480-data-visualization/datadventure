```python
import pandas as pd

# read the csv file
df = pd.read_csv("data/mxmh_survey_results.csv")

# look at the distribution of the depression column
print(df["Depression"].describe())
print(df["Depression"].value_counts().sort_index())

```

```python
# compute overall mean
mean_dep = df['Depression'].mean()

# find one high (max) and one low (min)
high = df.loc[df['Depression'].idxmax()]
low  = df.loc[df['Depression'].idxmin()]

# find one whose depression is closest to the mean
df['dist_to_mean'] = (df['Depression'] - mean_dep).abs()
mid  = df.loc[df['dist_to_mean'].idxmin()]

# collect them
examples = [low, mid, high]

# pick the columns you actually need for your bubble plot
# (you said: Hours per day, Loudness, Tempo, Danceability, Energy)
# adjust these names to match your real columns
cols = ['Frequency [Country]', 'Frequency [K pop]', 'Frequency [Pop]', 'Frequency [R&B]', 'Frequency [Video game music]', 'Depression', 'Fav genre', ]
summary = pd.DataFrame(examples)[cols]

print(summary.to_string(index=False))
```
